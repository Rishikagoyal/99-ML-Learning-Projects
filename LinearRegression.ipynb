```python
import urllib.request
import nltk
import string
nltk.download('punkt')
nltk.download('popular', quiet=True)
nltk.download('wordnet')
import warnings
warnings.filterwarnings('ignore')
from nltk.stem import WordNetLemmatizer
import random
import numpy as np



#write your code here
# Step 1: Import necessary libraries
import nltk
import re
import string
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Step 2: Read data from the .txt file and store in 'raw'
with open('https://cleveredailabs.s3.ap-south-1.amazonaws.com/india_192.txt', 'r', encoding='utf-8') as file:
  raw = file.read()

# Step 3: Text processing functions

# Lowercase the text
raw = raw.lower()

# Sentence tokenization
sentences = sent_tokenize(raw)

# Initialize lemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to clean and tokenize text
def clean_text(text):
# Remove punctuation
  text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)
# Tokenize the text
  tokens = word_tokenize(text)
# Lemmatize and remove stopwords
  tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
  return tokens

# Process the raw text
processed_sentences = [' '.join(clean_text(sentence)) for sentence in sentences]

# Step 4: Function to handle user queries using TF-IDF and cosine similarity
def response(user_input):
# Step 4.1: Add user input to the list of sentences
  sentences_with_input = processed_sentences + [user_input]

# Step 4.2: Vectorize the sentences using TF-IDF
  vectorizer = TfidfVectorizer()
  tfidf_matrix = vectorizer.fit_transform(sentences_with_input)

# Step 4.3: Calculate cosine similarity between user input and all other sentences
  cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1])

# Step 4.4: Find the sentence with the highest similarity score
  highest_sim_idx = np.argmax(cosine_similarities)

# Step 4.5: Return the most similar sentence as the response
  return sentences[highest_sim_idx]

# Step 5: Simulate conversation with the bot
print("Welcome to the India information bot!")
print("Ask me anything about the history, politics, economics, culture, or education of India.")

# Examples of how to call the response function with sample questions
user_question = "History of India"
print(f"\nYou: {user_question}")
bot_response = response(user_question.lower())
print(f"Bot: {bot_response}")

user_question = "Politics and government in India"
print(f"\nYou: {user_question}")
bot_response = response(user_question.lower())
print(f"Bot: {bot_response}")

user_question = "Economic relations of India"
print(f"\nYou: {user_question}")
bot_response = response(user_question.lower())
print(f"Bot: {bot_response}")

user_question = "Cultures in India"
print(f"\nYou: {user_question}")
bot_response = response(user_question.lower())
print(f"Bot: {bot_response}")

user_question = "Education system in India"
print(f"\nYou: {user_question}")
bot_response = response(user_question.lower())
print(f"Bot: {bot_response}")
```

```python
# importing libraries required
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import urllib.request

#import data
url = 'https://confrecordings.ams3.cdn.digitaloceanspaces.com/text_file.txt'
file = urllib.request.urlopen(url)
#Remove '#' before below line to see file data and put again '#' symbol.

raw=file.read().decode("utf-8")
print(raw)

# Create a TF-IDF vectorizer object
vectorizer = TfidfVectorizer()

# Calculate the TF-IDF scores for each word
tfidf_scores = vectorizer.fit_transform([raw])

# Create a Pandas DataFrame to store the TF-IDF scores
df = pd.DataFrame(tfidf_scores.T.todense(), index=vectorizer.get_feature_names_out(), columns=["TF-IDF"])

# Sort the words by their TF-IDF scores
df = df.sort_values('TF-IDF', ascending=False)

# Create a word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white', min_font_size=10).generate_from_frequencies(df['TF-IDF'])

# Display the word cloud
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()

```
